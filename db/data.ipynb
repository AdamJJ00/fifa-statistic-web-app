```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URLs for the webpages we want to scrape
urls = [
    "https://www.transfermarkt.com/poland/kader/verein/3442/saison_id/2021/plus/1",
    "https://www.transfermarkt.com/argentina/kader/verein/3437/saison_id/2021/plus/1",
    "https://www.transfermarkt.com/mexico/kader/verein/6303/saison_id/2021/plus/1",
    "https://www.transfermarkt.com/saudi-arabia/kader/verein/3807/saison_id/2021/plus/1",
    "https://www.transfermarkt.com/japan/kader/verein/3435/saison_id/2021/plus/1",
    "https://www.transfermarkt.com/spain/kader/verein/3375/saison_id/2021/plus/1",
    "https://www.transfermarkt.com/germany/kader/verein/3262/saison_id/2021/plus/1",
    "https://www.transfermarkt.com/costa-rica/kader/verein/8497/saison_id/2021/plus/1"
]

# Headers to send with the request
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}

# Initialize an empty list to store the player data
players = []

# Loop through each URL
for url in urls:

    # Make a GET request to the webpage
    response = requests.get(url, headers=headers)

    # Parse the HTML content of the webpage
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all of the table rows in the HTML
    rows = soup.find_all('tr')


    # Loop through each table row
    for row in rows:
        # Find all of the table cells in the row
        cells = row.find_all('td')
        # If there are at least 7 cells, then we know it's a player row
        if len(cells) >= 7:
            name = cells[3].text.strip()
            dob = cells[5].text.strip()
            height = cells[7].text.strip()
            position = cells[4].text.strip()
            market_value = cells[12].text.strip()
            better_foot = cells[8].text.strip()
            team = url.split('/')[3].capitalize()

            # Add a dictionary containing the player data to the list
            players.append({'Name': name, 'Date_of_birth': dob, 'Height_cm': height, 'Position': position, 'Market_value': market_value, 'Better_foot': better_foot, 'Team': team})

# Create a Pandas DataFrame from the list of player dictionaries
df = pd.DataFrame(players)
df.to_csv('players.csv')

# Print the DataFrame
df
```
